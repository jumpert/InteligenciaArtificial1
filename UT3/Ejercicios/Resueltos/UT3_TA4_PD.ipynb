### Parte 1 - Carga de datos y Preparación

```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

Leer el archivo `Cardiac Training`

```python
input_file = "src/cardiac-training.csv"
df = pd.read_csv(input_file, header=0)
print(df.values)
```

```python
print(df.head())
```

```python
X = df.loc[:, df.columns != '2do_Ataque_Corazon']
y = df['2do_Ataque_Corazon'].values
```

### Parte 2 - Entrenamiento y testing

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.30, random_state=
0, shuffle=True)
```

-Analizar cómo se utiliza el método de Logistic Regression y documentarlo (alternativas, parámetros). Entregar el documento.

El método de Regresion Logística es un algoritmo de clasificación que se utiliza para predecir la probabilidad de una variable pertenezca a una clase u otra. En la regresión logística, la variable de respuesta es binaria o categórica, es decir, solo contiene datos codificados como 0-1 o Si-No. En otras palabras, la regresión logística modela la probabilidad de que la variable de respuesta sea 1 o 0 para diferentes valores de entrada.

La forma de la función logística es la siguiente:

$$ f(x) = \frac{1}{1+e^{-x}} $$

La función logística es una función sigmoidea, toma un valor real como entrada y produce un valor entre 0 y 1 como salida. La función logística se puede utilizar para convertir cualquier valor real entre 0 y 1.

La regresión logística utiliza una función logística para modelar una variable de respuesta binaria. La función logística se puede escribir como:

$$ f(x) = \frac{1}{1+e^{-\beta x}} $$

$$ f(x) = \frac{e^{\beta x}}{1+e^{\beta x}} $$

$$ f(x) = \frac{1}{1+e^{-\beta_0 - \beta_1 x_1 - \beta_2 x_2 - ... - \beta_n x_n}} $$

Donde $\beta_0, \beta_1, \beta_2, ..., \beta_n$ son los coeficientes del modelo y $x_1, x_2, ..., x_n$ son las variables de entrada.


```python
lr = LogisticRegression()

# Ajustar el modelo. Evitar el error de convergencia aumentando el número de iteraciones
# Incrementar el número de iteraciones a 100000

lr = LogisticRegression(max_iter=100000)
lr = lr.fit(train_X, train_y)
```

### Parte 3 - Evaluación

```python
y_pred = lr.predict(test_X)
print("Predicted vs Expected")
print(y_pred)
print(test_y)
```

```python
print(classification_report(test_y, y_pred, digits=3))
```

```python
print(confusion_matrix(test_y, y_pred))
```
